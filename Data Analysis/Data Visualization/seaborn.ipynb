```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Visualization on the basis of two values
1.categorical data
2.numerical data

Basic Graph Types
1.Line Graph : show trends over time
2.Bar Graph : compare values between categories(discrete data)
3.Pie Chart: compare values between catefories(percentage data)
4.Histogram : show distribution of numerical data
5.Box plot: statistical distribution of numerical data
6.Heat map: show matrix of values using color
7.Scatter plot: show relationship between two variables
8.Bubble chart: show relationship between three variables

```python
df=sns.load_dataset('titanic')
df
```

```python
#count plot : survival and not survived
sns.countplot(df,x='survived',hue='sex',palette='hls')  #hls,RdBu =palette names
plt.title('Survived vs Non survived based on gender')
plt.show()
```

```python
sns.barplot(df,x='sex',y='fare',hue='class')
plt.title('Average fare between gender based on class')
```

```python
sns.pointplot(df,x='class',y='survived',hue='sex',markers=['o','x'])
plt.title('Survival Rate by Class and Gender')
plt.show()
```
